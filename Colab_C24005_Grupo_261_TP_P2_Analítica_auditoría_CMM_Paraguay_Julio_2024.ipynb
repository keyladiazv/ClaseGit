**Modelado**

```python
#Renombrar la columna distributor de sales_in_paraguay por id_distributor y la columna id de distributors_profile por id_distributor
df_sales_in_Paraguay.rename(columns={'distributor': 'id_distributor'}, inplace=True)
df_distributors_profile.rename(columns={'id': 'id_distributor'}, inplace=True)
```

```python
# Realizar merge con las columnas id_distributor y distributor
df_merged = pd.merge(df_sales_in_Paraguay, df_distributors_profile[['id_distributor', 'distributor']], on='id_distributor', how='inner')
```

```python
# Obtener el orden de las columnas actuales
columnas = list(df_merged.columns)
# Mover la columna 'distributor' al lado de 'id_distributor'
columnas.remove('distributor')
index_id_distributor = columnas.index('id_distributor') + 1
columnas.insert(index_id_distributor, 'distributor')
df_merged = df_merged[columnas]
df_sales_in_Paraguay = df_merged
```

```python
# Ocultar id_distributor
columna_a_ocultar = 'id_distributor'
df_sales_in_Paraguay_sin_id = df_sales_in_Paraguay.drop(columna_a_ocultar, axis=1)
df_sales_in_Paraguay = df_sales_in_Paraguay_sin_id
#Dar formato
pd.options.display.float_format = '{:,.2f}'.format
df_sales_in_Paraguay
```

```python
# Crear una copia explícita del df_exports_to_Paraguay para evitar SettingWithCopyWarning
df_exports_to_Paraguay = df_exports_to_Paraguay.copy()
# Renombrar la columna 'distributor' a 'id_distributor' en la copia del DataFrame
df_exports_to_Paraguay.rename(columns={'distributor': 'id_distributor'}, inplace=True)
```

```python
# Realizar merge con las columnas id_distributor y distributor
df_merged_exports = pd.merge(df_exports_to_Paraguay, df_distributors_profile[['id_distributor', 'distributor']], on='id_distributor', how='inner')
```

```python
# Obtener el orden de las columnas actuales
columnas_exports = list(df_merged_exports.columns)
# Mover la columna 'distributor' al lado de 'id_distributor'
columnas.remove('distributor')
index_id_distributor = columnas.index('id_distributor') + 1
columnas.insert(index_id_distributor, 'distributor')
df_merged_exports = df_merged_exports[columnas]
df_exports_to_Paraguay = df_merged_exports
```

```python
# Ocultar id_distributor
columna_a_ocultar_exports = 'id_distributor'
df_exports_to_Paraguay_sin_id = df_exports_to_Paraguay.drop(columna_a_ocultar_exports, axis=1)
df_exports_to_Paraguay = df_exports_to_Paraguay_sin_id
df_exports_to_Paraguay
```

```python
# Crear un nuevo DataFrame con los totales de ventas por producto para sales_in_Paraguay
ventas_totales = df_sales_in_Paraguay.iloc[:,1:].sum().sort_values(ascending=False)
ventas_totales = pd.DataFrame(ventas_totales, columns = ["Total_Ventas"])
ventas_totales
```

```python
# Crear un nuevo DataFrame con los totales de ventas por producto para exports_to_Paraguay
exports_totales = df_exports_to_Paraguay.iloc[:,1:].sum().sort_values(ascending=False)
exports_totales = pd.DataFrame(exports_totales, columns = ["Total_Exportación"])
exports_totales
```

**Análisis**

Se observa que las exportaciones de Steel y Brick no se alinean con las ventas locales. Por lo tanto, se recomienda llevar a cabo una evaluación para identificar las razones detrás de las bajas ventas de estos productos.

**Visualización**

```python
# Total de Ventas por Producto
plt.figure(figsize=(12, 8))
plt.barh(ventas_totales.index, ventas_totales["Total_Ventas"], color=['#470f2c', '#092e2a', '#300425', '#261047', '#420f31', '#400a18', '#383004', '#141499', '#6b361b', '#2b180e', '#220636', '#570a32'])
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{x * 1e-6:.1f}M'))
plt.xlabel('Total de Ventas (millones)')
plt.ylabel('Productos')
plt.title('Total de Ventas por Producto')
plt.show()
```

```python
# Total de Exportación por Producto
plt.figure(figsize=(12, 8))
plt.barh(exports_totales.index, exports_totales["Total_Exportación"], color=['#3f1d6e', '#171063', '#393752', '#34333d', '#323d45', '#274744', '#1e4a33', '#2b522d', '#446337', '#4b522c', '#545027', '#381a13'])
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{x * 1e-6:.1f}M'))
plt.xlabel('Total Exportación (millones)')
plt.ylabel('Productos')
plt.title('Total de Exportación por Producto')
plt.show()
```
